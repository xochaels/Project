## Matplotlib

### Import Library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Fungsi plt.figure() dan plt.title()

```python
plt.figure(figsize=(10,5))
plt.title('Ini adalah Judul')
plt.show()
```

### 2. Fungsi plt.xlabel()

```python
plt.figure(figsize=(10,5))
plt.title('Ini adalah Judul')

plt.xlabel('ini adalah x label')

plt.show()
```

### 3. Fungsi plt.ylabel()

```python
plt.figure(figsize=(10,5))
plt.title('Ini adalah Judul')
plt.xlabel('ini adalah x label')

plt.ylabel('ini adalah y label')

plt.show()
```

### 4. Fungsi plt.bar()

```python
df['ANGKATAN'].value_counts() #mengihung jumlah data yang sama
```

```python
labels = '2021','2020'

plt.figure(figsize=(10,5))
plt.title('Perbandingan Jumlah Mahasiswa BISNIS DIGITAL ANGKATAN 2020 DAN 2021')
plt.xlabel('ANGKATAN')
plt.ylabel('MAHASISWA')

plt.bar(labels, df['ANGKATAN'].value_counts() )
plt.show()
```

### 5. Fungsi plt.li()

```python
plt.title('Perbandingan Jumlah Mahasiswa BISNIS DIGITAL ANGKATAN 2020 DAN 2021')
plt.pie(df['ANGKATAN'].value_counts(), labels=labels, autopct='%1.1f%%')
plt.show()
```
